Data Preparation

```python
import pandas as pd
import numpy as np
import seaborn as sns


```

Importing Data


```python
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url)
df.head()
```

Getting to Know my Datashee

```python
#describe my data
df.describe()
```

```python
#shape my data set
df.shape
```

Handling Duplicates

```python
#check duplicated
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

Handling Null Values

```python
#check null values
df.isnull().sum()
```

Dropping Nulls

```python
#drop nulls
df.dropna(inplace=True)
#dropping a specific column
df1 = df.dropna(subset=['total_cases'])
```

Imputation

```python
#Numerical Values
#using mean
df['total_cases'] = df['total_cases'].fillna(df['total_cases'].mean())
#using median
df['total_cases'] = df['total_cases'].fillna(df['total_cases'].median())
#Now use median to impute null values
```

```python
#Categorical Values
#Select any Column with Categorical Values, impute using mode
df['continent'] = df['continent'].fillna(df['continent'].mode())


```

Dropping Columns

```python
#drop column named iso_code
df.drop('iso_code', axis=1, inplace=True)
#drop column named new_deaths_smoothed
df.drop('new_deaths_smoothed', axis=1, inplace=True )
```

Creating a subset

```python


#subset of data for Zimbabwe and Rwanda
subset = df[(df['location'] == 'Zimbabwe') | (df['location'] == 'Rwanda')]
#displaying the first few rows of subset
subset.head()
```

Let's save our cleaned up data

```python
#saving the data
df.to_csv('url')
```
